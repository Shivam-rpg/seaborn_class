# Seaborn – Class 2: Categorical Plots


## What Are Categorical Plots?

Categorical plots visualize data where at least one variable is **categorical**  
(e.g., day of week, gender, class).

Seaborn gives several easy options to:
- Compare **counts** or **averages** across categories
- Understand spread & distribution


## barplot() vs countplot()

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", data=tips) # barplot - show relation in category at x and continuous data at y, and black line show variance
```

```python
sns.countplot(x="day", data=tips) # countplot - show count of each category
```

## boxplot() & violinplot()

```python
sns.boxplot(x="day", y="total_bill", data=tips) # boxplot - show spread & distribution and o - shows the outliers
```

```python
sns.violinplot(x="day", y="total_bill", data=tips) # violinplot - show spread & distribution
```

## stripplot() & swarmplot()

```python
sns.stripplot(x="day", y="total_bill", data=tips) # stripplot - show every observation
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips) # swarmplot - show every observation
```

| Plot        | Best for |
|-------------|-----------|
| **barplot** | Compare average values across categories |
| **countplot** | Show how many items fall in each category |
| **boxplot** | See median, quartiles, and outliers |
| **violinplot** | Same as boxplot + shows distribution |
| **stripplot** | View every observation (good for small data) |
| **swarmplot** | Like stripplot but avoids overlap |


```python
sns.swarmplot(x="day", y="total_bill", data=tips)
sns.violinplot(x="day", y="total_bill", data=tips)
```
