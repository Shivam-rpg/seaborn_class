## Why Matrix / Grid Plots?

Matrix or grid-style plots are perfect for:
- Showing correlations between variables  
- Visualizing pivot tables  
- Discovering patterns in large datasets


### Heatmaps (heatmap())

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
tips.head()

sns.heatmap(tips.corr(numeric_only= True), annot=True)
```

### Cluster Map (clustermap())

```python
iris = sns.load_dataset("iris") # mostly used in Machine Learning Unsuperwised learning
iris.head()

sns.clustermap(iris.corr(numeric_only= True), annot=True, cmap="coolwarm")
```
